# 이미지 테두리 추출 코드 설명

## 코드 기능

코드는 OpenCV 라이브러리를 사용하여 이미지 테두리를 추출하는 기능을 구현합니다.주요 단계는 다음과 같습니다.

1. 그림 읽기
2. 그림을 그레이스케일로 변환하기
3. Canny 에지 탐지 사용
4. 그림의 윤곽을 찾다
5. 그림의 테두리를 그리고 결과를 표시합니다

## 코드 사용

'extract_image_border' 함수를 호출하여 이미지 경로를 불러오면 테두리가 있는 이미지를 표시할 수 있습니다.

```python
image_path = 'path/to/your/image.jpg'
extract_image_border(image_path)

````python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

def extract_image_border(image_path):
    # 이미지 읽기
    img = cv2.imread(image_path)

    # 그레이스케일로 변환
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Canny 에지 검출 사용
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)

    # 윤곽선 찾기
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 빈 이미지 생성
    border_img = np.zeros_like(img)

    # 윤곽선 그리기
    cv2.drawContours(border_img, contours, -1, (0, 255, 0), 2)

    # 결과 표시
    cv2_imshow(border_img)

# 함수 호출 및 이미지 경로 전달
image_path = '/content/drive/MyDrive/computer vision/CU_LOGO.jpg'
extract_image_border(image_path)
````
